**2. Function `find_product`**

**Instructions**

   - Complete the function `find_product` that takes four parameters - **dataframe,state,year and month**
   - Function Should return the **Category** which had the most number orders in that **State** during the specified **Month and Year**
   - Return as **String**
   - Use Groupby
   - Use Column Order Date


```python
def find_product(dataframe,state,year,month):
    
    orders_category = dataframe.loc[dataframe['State']==state]
    orders_category = orders_category.loc[orders_category['Order Date'].dt.year==year]
    orders_category = orders_category.loc[orders_category['Order Date'].dt.month==month]
    
    orders_category = orders_category.groupby(by='Category').max().reset_index()['Category'][0]
    orders_category
    return orders_category
#     #End

q2=[find_product(df,'Florida',2017,3),find_product(df,'Ohio',2017,5)]
print(q2)
```

**3. Write the code to create a pivot table that contains average sales of each month in each year.**

**Instructions**

- Index should be **Month** and columns must be **Year**
- Month and Year should be in ascending order
- Hint use `pd.pivot_table`
- Use `np.mean`

**Sample Output**

<img src="Sample/sample1.PNG" >

```python
#Enter code below

dup1 = pd.DataFrame(df, columns=['Sales','Month'])
dup1['Year'] = df['Ship Date'].dt.year
dup1['Month'] = df['Ship Date'].dt.month

dup1

table =  pd.pivot_table(dup1, values ='Sales', index =['Month'],
                         columns =['Year'], aggfunc = (np.mean))

q3=table.round(2)
q3
#End
# print(q3)
```
